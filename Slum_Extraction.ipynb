###Nairobi

```python
import os, glob, shutil
import rasterio
import numpy as np

# Original Nairobi folder that contains test, train, val
base_folder = "/content/drive/MyDrive/IDEABENCH_Project /v2/nairobi_kenya"

# New output folder where extracted RF TIFFs will be copied
output_folder = "/content/drive/MyDrive/IDEABENCH_Project /nairobi_kenya_RF_value_2"

os.makedirs(output_folder, exist_ok=True)

total_copied = 0

for split in ["test", "train", "val"]:
    split_folder = os.path.join(base_folder, split)

    rf_files = glob.glob(os.path.join(split_folder, "RF_*.tif"))

    print(f"{split}: RF files found = {len(rf_files)}")

    for file in rf_files:
        with rasterio.open(file) as src:
            data = src.read(1)

            if src.nodata is not None:
                data = data[data != src.nodata]

            unique_values = np.unique(data)

        if 2 in unique_values:
            # keep split name in filename so test/train/val files don't overwrite each other
            new_name = f"{split}_{os.path.basename(file)}"
            destination = os.path.join(output_folder, new_name)

            shutil.copy(file, destination)
            total_copied += 1

print("Total RF TIFFs copied with value 2:", total_copied)
print("Saved in:", output_folder)
```
